```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
dogecoin_data = cg.get_coin_market_chart_by_id(id='dogecoin',vs_currency = 'usd',days=30)
print(dogecoin_data)
```

```python
import pandas as pd
dogecoin_data_final = pd.DataFrame(dogecoin_data['prices'],columns=['TimeStamp','Price'])
dogecoin_data_final['Date']=pd.to_datetime(dogecoin_data_final['TimeStamp'],unit='ms')
dogecoin_data_final.head()
```

```python
candlestick_data_dogecoin = dogecoin_data_final.groupby(dogecoin_data_final.Date.dt.date).agg({'Price':['min','max','first','last']})
candlestick_data_dogecoin.head()
```

```python
import plotly.graph_objects as go
from plotly import offline
fig = go.Figure(data=[go.Candlestick(x=candlestick_data_dogecoin.index,
                open = candlestick_data_dogecoin['Price']['first'],
                high = candlestick_data_dogecoin['Price']['max'],
                low = candlestick_data_dogecoin['Price']['min'],
                close = candlestick_data_dogecoin['Price']['last'])
                ])
fig.update_layout(xaxis_rangeslider_visible=False,xaxis_title='Date',yaxis_title='Price (USD $)', title = 'Dogecoin Candlestick Chart Over Past 30 Days')
offline.plot(fig,filename='dogecoin_candlestick_graph.html')
```

```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
cg = CoinGeckoAPI()
bitcoin_data = cg.get_coin_market_chart_by_id(id='bitcoin',vs_currency = 'usd',days=30)
print(bitcoin_data)
```

```python
bitcoin_data_final = pd.DataFrame(bitcoin_data['prices'],columns=['TimeStamp','Price'])
bitcoin_data_final['Date']=pd.to_datetime(bitcoin_data_final['TimeStamp'],unit='ms')
bitcoin_data_final.head()
```

```python
candlestick_data_bitcoin = bitcoin_data_final.groupby(bitcoin_data_final.Date.dt.date).agg({'Price':['min','max','first','last']})
candlestick_data_bitcoin.head()
```

```python
import plotly.graph_objects as go
from plotly import offline
fig = go.Figure(data=[go.Candlestick(x=candlestick_data_bitcoin.index,
                open = candlestick_data_bitcoin['Price']['first'],
                high = candlestick_data_bitcoin['Price']['max'],
                low = candlestick_data_bitcoin['Price']['min'],
                close = candlestick_data_bitcoin['Price']['last'])
                ])
fig.update_layout(xaxis_rangeslider_visible=False,xaxis_title='Date',yaxis_title='Price (USD $)', title = 'Bitcoin Candlestick Chart Over Past 30 Days')
offline.plot(fig,filename='bitcoin_candlestick_graph.html')
```

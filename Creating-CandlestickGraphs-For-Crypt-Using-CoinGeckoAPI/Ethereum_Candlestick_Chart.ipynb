```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
ethereum_data = cg.get_coin_market_chart_by_id(id='ethereum',vs_currency = 'usd',days=30)
print(ethereum_data)
```

```python
import pandas as pd
ethereum_data_final = pd.DataFrame(ethereum_data['prices'],columns=['TimeStamp','Price'])
ethereum_data_final['Date']=pd.to_datetime(ethereum_data_final['TimeStamp'],unit='ms')
ethereum_data_final.head()
```

```python
candlestick_data_ethereum = ethereum_data_final.groupby(ethereum_data_final.Date.dt.date).agg({'Price':['min','max','first','last']})
candlestick_data_ethereum.head()
```

```python
import plotly.graph_objects as go
from plotly import offline
fig = go.Figure(data=[go.Candlestick(x=candlestick_data_ethereum.index,
                open = candlestick_data_ethereum['Price']['first'],
                high = candlestick_data_ethereum['Price']['max'],
                low = candlestick_data_ethereum['Price']['min'],
                close = candlestick_data_ethereum['Price']['last'])
                ])
fig.update_layout(xaxis_rangeslider_visible=False,xaxis_title='Date',yaxis_title='Price (USD $)', title = 'Ethereum Candlestick Chart Over Past 30 Days')
offline.plot(fig,filename='ethereum_candlestick_graph.html')
```
